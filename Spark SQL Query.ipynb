```python
import pyspark

from pyspark.sql import SparkSession
from pyspark.context import SparkContext 
sc = SparkSession.builder.master("local[*]").config("spark.mongodb.input.uri","mongodb://localhost:27017/Dbda.aj").config("spark.mongodb.output.uri","mongodb://localhost:27017/Dbda.aj").config("spark.jars.packages","org.mongodb.spark:mongo-spark-connector_2.12:3.0.0").getOrCreate()
```

```python
from pyspark.sql.functions import *

from pyspark import SparkFiles

df=sc.read.options(inferSchema='True',delimiter=',',header='True').csv("startup.csv")
df1=sc.read.options(inferSchema='True',delimiter=',',header='True').parquet("consumerInternet.parquet")
df.show(5)
df.count()
```

```python
df1.show(5)
df1.count()
```

```python
distinct_ids = [x.City for x in df.select('City').distinct().collect()]
distinct_ids
```

```python
distinct_ids = [x.Startup_Name for x in df1.select('Startup_Name').distinct().collect()]
distinct_ids
```

```python
df3 = df.union(df1)
df3.show(truncate=False)
df3.count()
```

```python
## How many startups are there in Pune City?

df3.select(df3.Startup_Name).filter(df3.City == 'Pune').count()
```

```python
##  How many startups in Pune got their Seed/ Angel Funding?

df3.select(df3.Startup_Name).filter((df3.City == 'Pune') & (df3.InvestmentnType.isin('Seed/ Angel Funding','Seed/Angel Funding','Seed / Angle Funding'))).count()
```

```python
## What is the total amount raised by startups in Pune City? Hint - use regex_replace to get rid of null

df4 = df3.withColumn("Amount_in_USD",regexp_replace("Amount_in_USD", "[^0-9]", ""))
df3 = df4.withColumn("Amount_in_USD",df4["Amount_in_USD"].cast("int"))
df3.filter(df3.City == 'Pune').groupby('City').sum('Amount_in_USD').show()
```
